```python
#Programacion dinamica
#Coeficiente binomial
import matplotlib.pyplot as plt
import time
import numpy as np
from math import factorial

def b2(n,k,tiempos):
  inicio = time.time()
  c=[]
  for i in range(n+1):
    c.append([])
    for j in range(k+1):
      c[i].append(0)

  for i in range (n+1):
    for j in range (min(i,k)+1):
      if j==0 or j==i:
        c[i][j]=1

      else:
        c[i][j]=c[i-1][j-1]+c[i-1][j]
  fin = time.time()
  tiempos.append(fin-inicio)
  return c[n][k]

  
tiempos = list()
numeros =list()
n=50
for i in range (n):
  numeros.append(b2(i+5,i+2,tiempos))  
  
plt.plot(numeros,tiempos)
plt.xlabel('coeficiente binomial')
plt.ylabel('tiempo')
plt.title('Lab DLS')
plt.show()
    
```

```python
#coeficiente binomial recursivo
import matplotlib.pyplot as plt
import time

def binomialCoeff(n, k):
  
  if k > n:
    return 0
  if k == 0 or k == n:
    return 1
#llamada recursiva
  
  return binomialCoeff(n-1, k-1) + binomialCoeff(n-1,k)

n=50
numeros =list()
tiempos = list()
for i in range (n):
  inicio = time.time()
  numeros.append(binomialCoeff(i+5,i+2))
  fin = time.time()
  tiempos.append(fin-inicio)
print(numeros)
plt.plot(numeros,tiempos)
plt.xlabel('coeficiente binomial')
plt.ylabel('tiempo')
plt.title('Lab DLS')
plt.show()
```
